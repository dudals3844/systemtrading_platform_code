```python
import pandas as pd
import numpy as np
```

```python
def calculate_alpha_month(price_df):
    try:
        if len(price_df) > 31:
            tmp_list = []
            for i in range(0, 30):
                today_start = price_df['시가'].iloc[i]
                today_end = price_df['종가'].iloc[i]
                today_high = price_df['고가'].iloc[i]
                yesterday_high = price_df['고가'].iloc[i+1]
                yesterday_low = price_df['저가'].iloc[i+1]
                spread = yesterday_high - yesterday_low
                standard = today_start + spread * 0.5
                if today_high > standard and today_start != 0:
                    if (today_high - today_end)/today_start > 0.02:
                        print((today_high - today_end)/today_start)
                        print(" %s %s %s" %(today_high, today_end, today_start))
                        profit = today_high*0.98 - standard
                        alpha = profit/today_start
                        tmp_list.append(alpha+1)
                    
                    else:
                        profit = today_end - standard
                        alpha = profit/today_start
                        tmp_list.append(alpha+1)


            alpha_array = np.array(tmp_list)
            total_alpha = np.prod(alpha_array) - 1
            return total_alpha


        else:
            print('상장된지 얼마 안됨')
            return 0

    except:
        pass
```

```python
def read_company_price(company_name):
    try:
        price_df = pd.read_excel('C:/Users/PC/PycharmProjects/systemtrading_platform/db/price/'+company_name+'.xlsx')
        return price_df
    except:
        pass
```

```python
def standard_code(code):
    code = "{0:0>6}".format(code)  # 오류안나게 종목코드 6자리 맞춰줌
    return code
```

```python
company_df = pd.read_excel("C:/Users/PC/PycharmProjects/systemtrading_platform/db/kosdaq상장회사.xls")
tmp_df = pd.read_excel("C:/Users/PC/PycharmProjects/systemtrading_platform/db/kospi상장회사.xls")
company_df = pd.concat([company_df, tmp_df], axis=0)
```

```python
alpha_list = []
for i in range(len(company_df)):
    company_name = company_df['기업명'].iloc[i]
    code = company_df['종목코드'].iloc[i]
    code = standard_code(code)
    price_df = read_company_price(company_name)
    total_alpha = calculate_alpha_month(price_df)
    alpha_list.append([code, company_name, total_alpha])
    print(code,' ',company_name, total_alpha)

all_company_alpha_df = pd.DataFrame(alpha_list, columns=['종목코드','기업명','수익률'])
all_company_alpha_df.sort_values(by=['수익률'], axis=0, ascending=False, inplace=True)
all_company_alpha_df.to_csv("C:/Users/PC/PycharmProjects/systemtrading_platform/db/손절변동성돌파전략수익률.csv")
```
